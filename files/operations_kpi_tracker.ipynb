This notebook tracks 12 months of operational KPIs for a large-scale program (modeled after real experience managing a $500K annual budget and 12,000+ user cafeteria operations program at Airstreams Renewables).

**Metrics tracked:**
- Compliance score trend
- Manual task reduction through workflow automation
- Budget vs. actual spend
- Daily user volume (demand planning)
- User satisfaction score

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import warnings
warnings.filterwarnings('ignore')

# Brand palette
GREEN = '#1B4332'
GOLD  = '#B8933A'
MUTED = '#4A5568'
BG    = '#F7F6F1'

print('Libraries loaded ✓')
```

## 1. Load & Prepare Data

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

data = {
    'Month':          months,
    'Daily_Users':    [285, 290, 310, 305, 320, 315, 298, 312, 328, 335, 342, 338],
    'Budget_Spent':   [38500, 39200, 41000, 40500, 42000, 41800, 39900, 41200, 43000, 44100, 45000, 44500],
    'Budget_Target':  [41000] * 12,
    'Compliance':     [72, 78, 82, 85, 87, 90, 88, 91, 93, 94, 96, 97],
    'Manual_Tasks':   [142, 135, 118, 102, 89, 76, 68, 62, 58, 54, 52, 50],
    'Waste_Pct':      [8.2, 7.8, 7.1, 6.8, 6.2, 5.9, 5.7, 5.4, 5.1, 4.8, 4.5, 4.3],
    'Satisfaction':   [3.6, 3.7, 3.8, 3.9, 4.0, 4.1, 4.1, 4.2, 4.3, 4.4, 4.5, 4.6]
}

df = pd.DataFrame(data)

# Derived metrics
df['Budget_Variance']  = df['Budget_Target'] - df['Budget_Spent']
df['Efficiency_Gain']  = ((df['Manual_Tasks'].iloc[0] - df['Manual_Tasks']) / df['Manual_Tasks'].iloc[0] * 100).round(1)
df['Compliance_Delta'] = df['Compliance'] - df['Compliance'].iloc[0]

df.head()
```

## 2. Executive KPI Summary

```python
summary = {
    'Avg Daily Users':           f"{df['Daily_Users'].mean():.0f}",
    'Total Budget Savings':      f"${df['Budget_Variance'].sum():,.0f}",
    'Compliance Improvement':    f"+{df['Compliance'].iloc[-1] - df['Compliance'].iloc[0]} pts",
    'Manual Task Reduction':     f"{df['Efficiency_Gain'].iloc[-1]}%",
    'Waste Reduction':           f"{df['Waste_Pct'].iloc[0] - df['Waste_Pct'].iloc[-1]:.1f} pp",
    'Satisfaction Improvement':  f"+{df['Satisfaction'].iloc[-1] - df['Satisfaction'].iloc[0]:.1f} pts",
    'Peak Month Users':          f"{df['Daily_Users'].max()} ({df.loc[df['Daily_Users'].idxmax(),'Month']})",
    'Best Compliance Month':     f"{df['Compliance'].max()} pts ({df.loc[df['Compliance'].idxmax(),'Month']})"
}

print('=' * 45)
print('   OPERATIONS PROGRAM — KPI SUMMARY')
print('=' * 45)
for k, v in summary.items():
    print(f"  {k:<28} {v}")
print('=' * 45)
```

## 3. KPI Dashboard — Visualization

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 8), facecolor=BG)
fig.suptitle('Operations KPI Dashboard — 12-Month Program Review',
             fontsize=14, fontweight='bold', color=GREEN, y=1.01)

x = range(12)

# — Compliance score
ax = axes[0, 0]
ax.set_facecolor('white')
ax.fill_between(x, df['Compliance'], alpha=0.12, color=GOLD)
ax.plot(x, df['Compliance'], color=GOLD, linewidth=2.5, marker='o', markersize=5,
        markerfacecolor=GREEN, markeredgecolor=GOLD, markeredgewidth=1.5)
ax.axhline(90, color=GREEN, linestyle='--', linewidth=1, alpha=0.5)
ax.set_xticks(x); ax.set_xticklabels(months, fontsize=8)
ax.set_title('Compliance Score', fontweight='bold', color=GREEN)
ax.set_ylabel('Score'); ax.set_ylim(65, 102)
ax.grid(axis='y', alpha=0.3)
for s in ['top','right']: ax.spines[s].set_visible(False)

# — Manual task reduction
ax = axes[0, 1]
ax.set_facecolor('white')
ax.bar(x, df['Manual_Tasks'], color=GREEN, alpha=0.75, width=0.6)
ax.plot(x, df['Manual_Tasks'], color=GOLD, linewidth=2, marker='o', markersize=4)
ax.set_xticks(x); ax.set_xticklabels(months, fontsize=8)
ax.set_title('Manual Task Reduction (Automation)', fontweight='bold', color=GREEN)
ax.set_ylabel('Tasks')
ax.grid(axis='y', alpha=0.3)
for s in ['top','right']: ax.spines[s].set_visible(False)

# — Budget vs target
ax = axes[1, 0]
ax.set_facecolor('white')
xn = np.arange(12)
ax.bar(xn - 0.2, [b/1000 for b in df['Budget_Target']], 0.38,
       label='Target', color='#CBD5E0', edgecolor='none')
ax.bar(xn + 0.2, [b/1000 for b in df['Budget_Spent']], 0.38,
       label='Actual',  color=GREEN, alpha=0.82, edgecolor='none')
ax.set_xticks(xn); ax.set_xticklabels(months, fontsize=8)
ax.set_title('Budget vs. Actual ($K)', fontweight='bold', color=GREEN)
ax.set_ylabel('$K')
ax.legend(fontsize=8, frameon=False)
ax.grid(axis='y', alpha=0.3)
for s in ['top','right']: ax.spines[s].set_visible(False)

# — Satisfaction
ax = axes[1, 1]
ax.set_facecolor('white')
bar_colors = [GOLD if s >= 4.0 else GREEN for s in df['Satisfaction']]
ax.bar(x, df['Satisfaction'], color=bar_colors, alpha=0.85, width=0.6)
ax.axhline(4.0, color=GOLD, linestyle='--', linewidth=1.2, alpha=0.7)
ax.set_xticks(x); ax.set_xticklabels(months, fontsize=8)
ax.set_title('User Satisfaction Score (/5.0)', fontweight='bold', color=GREEN)
ax.set_ylabel('Score')
ax.set_ylim(0, 5.2)
ax.grid(axis='y', alpha=0.3)
for s in ['top','right']: ax.spines[s].set_visible(False)

plt.tight_layout()
plt.show()
```

## 4. Trend Analysis & Correlation

```python
# Correlation matrix — does automation improvement correlate with compliance gain?
corr_cols = ['Compliance', 'Manual_Tasks', 'Waste_Pct', 'Satisfaction', 'Daily_Users']
corr = df[corr_cols].corr().round(2)

print('Correlation Matrix:')
print(corr.to_string())

print('\nKey insight:')
c = df['Manual_Tasks'].corr(df['Compliance'])
print(f'  Manual task volume vs Compliance score: r = {c:.2f}')
print('  → Reducing manual workload is strongly correlated with compliance improvement.')
```

## 5. Efficiency Gain Over Time

```python
fig, ax = plt.subplots(figsize=(10, 4), facecolor=BG)
ax.set_facecolor('white')

ax.fill_between(x, df['Efficiency_Gain'], alpha=0.15, color=GREEN)
ax.plot(x, df['Efficiency_Gain'], color=GREEN, linewidth=2.5, marker='o',
        markersize=6, markerfacecolor=GOLD, markeredgecolor=GREEN, markeredgewidth=2)

for i, (xi, yi) in enumerate(zip(x, df['Efficiency_Gain'])):
    if i in [0, 3, 6, 9, 11]:
        ax.annotate(f'{yi}%', xy=(xi, yi), xytext=(0, 10),
                    textcoords='offset points', ha='center',
                    fontsize=8.5, color=GREEN, fontweight='bold')

ax.set_xticks(x); ax.set_xticklabels(months)
ax.set_ylabel('Efficiency Gain (%)')
ax.set_title('Cumulative Efficiency Gain from Workflow Automation', fontsize=12,
             fontweight='bold', color=GREEN)
ax.set_ylim(-5, 80)
ax.grid(axis='y', alpha=0.3)
for s in ['top','right']: ax.spines[s].set_visible(False)

plt.tight_layout()
plt.show()

print(f"Final efficiency gain: {df['Efficiency_Gain'].iloc[-1]}% reduction in manual tasks")
```
